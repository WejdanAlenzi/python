```python
import numpy as np
a=np.array([[0.,0.,0.],
           [10.,10,10],
            [20,20,20],
            [30,30,30]])
b=np.array([0.,1.,2.])
c=a+b
print("c=\n",c)
mask=(c>=22)
print("\nmask=\n",mask)
count=mask.sum()
print("\ncount>=22:",count)
           
```

```python
import numpy as np

# إنشاء المصفوفة
grades = np.array([
    [87, 96, 70],
    [100, 87, 90],
    [94, 77, 90],
    [100, 81, 82]
])

# العمليات العامة على كامل المصفوفة
print("Overall:")
print("Sum:",grades.sum())
print("Min:",grades.min())
print("Max:",grades.max())
print("Mean:",grades.mean())
print("Std:",grades.std())
print("Var:",grades.var())

# المتوسط لكل عمود (axis=0)
print("\nmean(axis=0) per column:",grades.mean(axis=0))
# المتوسط لكل صف (axis=1)
print("Mean (axis=1) per row:",grades.mean(axis=1))

# القيمة العظمى لكل صف (axis=1)
print("\nmax(axis=1) per row:",grades.max(axis=1))
```

```python
import numpy as np
grades = np.array([[70,85,90],
                   [88,76,94],
                   [60,80,72]])
print("argmax overall:",grades.argmax())
print("argmax per row:",grades.argmax(axis=1))
print("argmin per column:",grades.argmin(axis=0))
```

```python
import numpy as np 
def manhattan_distance(x1,y1,x2,y2):
   return abs(x2 - x1) + abs(y2 - y1)
def euclidean_distance(x1,y1,x2,y2):
   return np.abs((x2 - x1)**2 + (y2 - y1)**2)
```
