```python
color = input("Enter a color (red, yellow, green): ")


if color == "":
    print("Invalid input")
else:
    #usr match
    match color:
        case "red":
            print("Stop")          
        case "yellow":
            print("Slow down")
        case "green":
            print("Go")            
        case _:
            print("Unknown color")
```

```python
# Displaying a list of mathematical operations
print("1) Addition")
print("2) Subtraction")
print("3) Multiplication")
print("4) Division")

choice = input("Choose an operation (1-4): ")

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

match choice:
    case "1":
        print("Result:", num1 + num2)
    case "2":
        print("Result:", num1 - num2)
    case "3":
        print("Result:", num1 * num2)
    case "4":
        if num2 == 0:
            print("Cannot divide by zero")
        else:
            print("Result:", num1 / num2)
    case _:
        print("Invalid choice")
```

```python
grade = float(input("Enter your score: "))

#pass or fail
if grade >= 60:
    print("Pass")
else:
    print("Fail")


if grade >= 90:
    letter = "A"
elif grade >= 80:
    letter = "B"
elif grade >= 70:
    letter = "C"
elif grade >= 60:
    letter = "D"
else:
    letter = "F"

print("Your grade is:", letter)


match letter:
    case "A":
        print("Excellent")
    case "B":
        print("Very Good")
    case "C":
        print("Good")
    case "D":
        print("Needs Improvement")
    case "F":
        print("Failing")
```
