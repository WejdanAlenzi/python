```python
try:
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    result = num1 / num2

except ZeroDivisionError:
    print("Error: Cannot divide by zero.")

except ValueError:
    print("Error: Please enter valid numbers only.")

else:
    print("Division result:", result)
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name          # public
        self._age = age           # private

    def set_age(self, new_age):
        self._age = new_age

    def display(self):
        print("Name =", self.name)
        print("Age =", self._age)


p = Person("Wejdan", 19)
p.display()
p.set_age(20)
print("After updating age:")
p.display()

```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary


class Manager(Employee):
    def add_bonus(self, bonus):
        self.salary += bonus

    def display(self):
        print("Manager Name:", self.name)
        print("Total Salary:", self.salary)


m = Manager("Wejdan", 5000)

m.add_bonus(1500)

m.display()
```

```python
class Student:
    def __init__(self):
        self.name = input("Enter your name: ")

        try:
            age_input = input("Enter your age: ")
            age = int(age_input)

            if age < 0:
                raise ValueError

            self.age = age

        except ValueError:
            print("Invalid age! Age must be a positive number.")
            self.age = None

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)


# Create object
s = Student()
s.display()
```
