```python
total = 0

for i in range(10): #counter
    num = int(input("Enter a number: "))

    if num < 0:
        continue

    if num == 0: #if the number is zero,we stop
        break

    if num % 2 == 0:
        total += num

print("final sum :", total)
```

```python
n = int(input("Enter a number: "))

for i in range(1, 13): #form 1 to 12
    result = n * i
#if i is a multiple of 3,we ignore it
    if i % 3 == 0:
        continue

    if result > 50:
        break

    print(n, "x", i, "=", result)
```

```python
secret = 12  
attempts = 0

while True: #endless loop
    guess = int(input("Guess a number between 1 and 20: "))

    if guess < 1 or guess > 20:
        continue

    attempts += 1

    if guess == secret:
        print("Correct! You guessed it in", attempts, "attempts.")
        break
```

```python
n = int(input("Enter a number: "))
count = 0

for i in range(1, n + 1):

    if i > 50:
        break

    if i % 8 == 0: #if it is divisible by 8, we ignore it
        continue

    if i % 4 == 0:
        count += 1

print("Count of numbers divisible by 4 is:", count)
```

```python
for i in range(1, 6): #1 to 5
    for j in range(1, i + 1):

        if j == 3:
            continue

        if i * j > 6: #if the product of i and j is greater than 6, we stop
            break

        print(j, end=" ")

    print()
```
