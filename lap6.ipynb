```python
# Ask the user to enter 5 numbers
numbers = []

for i in range(5):
    num = float(input(f"Enter number {i+1}: "))
    numbers.append(num)


total = sum(numbers)
average = total / len(numbers)
maximum = max(numbers)
minimum = min(numbers)

# Print results
print("Numbers:", numbers)
print("Sum:", total)
print("Average:", average)
print("Maximum:", maximum)
print("Minimum:", minimum)
```

```python
# Ask the user to enter 5 words
words = tuple(input(f"Enter word {i+1}: ") for i in range(5))


# Print first, last, and count
print("Words tuple:", words)
print("First word:", words[0])
print("Last word:", words[-1])
print("Total words:", len(words))

search_word = input("Enter a word search: ")

if search_word in words:
    print(f"'{search_word}' is in the tuple")
else:
    print(f"'{search_word}' is NOT in the tuple")
```

```python
students = {}

# Enter 3 students and scores
for i in range(3):
    name = input(f"Enter name of student {i+1} name: ")
    score = float(input(f"Enter {name}'s score: "))
    students[name] = score

print("Students scores:", students)
max_score = max(students.values())
top_students = [name for name, score in students.items() if score == max_score]

print("Top student(s):", top_students, "with score:", max_score)
```

```python
# Enter two lists of numbers
list1 =list(map(int, input("Enter numbers for list 1 (space-separated): ").split()))
list2 =list(map(int, input("Enter numbers for list 2 (space-separated): ").split()))


set1 = set(list1)
set2 = set(list2)

print("Set 1:", set1)
print("Set 2:", set2)     
print("Union:", set1 | set2)
print("Intersection:", set1 & set2)
print("Difference (set1 - set2):", set1 - set2)
```

```python
names_list = [input(f"Enter name {i+1}: ") for i in range(5)]


names_tuple = tuple(names_list)
print("Names tuple:", names_tuple)

names_dict = {name: len(name) for name in names_list}
print("Names dictionary :", names_dict)

length_set = set(names_dict.values())
print("Unique name lengths set:", length_set)
```
