```python
import os
os.makedirs("lab4_data/input",exist_ok=True)
os.makedirs("lab4_data/output",exist_ok=True)
os.makedirs("lab4_data/logs",exist_ok=True)
for item in os.listdir("lab4_data"):
    print(item)

os.rename("lab4_data/logs", "lab4_data/history")
```

```python
#Create and write to a text file
with open("lab4 data/input/notes.txt", "w") as f:
    f.write("Python is easy\n")
    f.write("Files are useful\n")
    f.write("This is Lab 4\n")
    f.write("We are learning file handling\n")
    f.write("Text files store information\n")

#Read file and count lines
with open("lab4 data/input/notes.txt", "r") as f:
    lines = f.readlines()
    print("Lines:", len(lines))

#Copy fil
with open("lab4 data/input/notes.txt", "r") as f:
  text = f.read()
with open("lab4 data/output/summary.txt", "w") as f:
    f.write(text)
```

```python
import pyfiglet
#Create ASCII text
ascii_text =  pyfiglet.figlet_format("WEJDAN") #replace (Your name)
print(ascii_text)
#Make sure the output folder exists
os.makedirs("lab4_data/output", exist_ok=True)
#Save ASCII text
output_file = "lab4_data/output/ascii_name.txt"
with open(output_file, "w") as f:
    f.write(ascii_text)
print(f"ASCII text saved to {output_file}") 
```
